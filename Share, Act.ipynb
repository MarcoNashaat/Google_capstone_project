```python
# improting libraries for data processing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# loading data from the CSV file into pandas dataframe
df = pd.read_csv('final_data.csv')
```

```python
fig, ax = plt.subplots() #create figure and axis object
ax.bar(df['user_type'].value_counts().index, df['user_type'].value_counts().values) #plot the bar chart

# set the title
ax.set_title('Subscribers are much more than Customers')

# show the chart
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(16,6)) #create figure and axis object


ax1.bar(df[df['user_type']=='Subscriber']['day_of_week'].value_counts().index, df[df['user_type']=='Subscriber']['day_of_week'].value_counts().values,width=.4) 

ax2.bar(df[df['user_type']=='Customer']['day_of_week'].value_counts().index, df[df['user_type']=='Customer']['day_of_week'].value_counts().values,width=.4) 

# set the title
fig.suptitle('User counts over days of the week, Subscribers[left] vs Customers[right]')

# show the chart
plt.show()
```

```python
fig, ax = plt.subplots() 

x = df.groupby('user_type').mean()
ax.bar(x.index, x['duration']/60)

# set the title
fig.suptitle('Average trip duration')
plt.ylabel('Duration in mins')

# show the chart
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(16,6)) #create figure and axis object

x = df[df['user_type']=='Subscriber']['start_station_name'].value_counts().iloc[:5]
y = df[df['user_type']=='Customer']['start_station_name'].value_counts().iloc[:5]
ax1.bar(x.index, x.values,width=.4) 
ax1.tick_params(axis='x', rotation=45)
ax2.bar(y.index, y.values,width=.4) 
ax2.tick_params(axis='x', rotation=45)
# set the title
fig.suptitle('Most popular start station, Subscribers[left] vs Customers[right]')

# show the chart
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(16,6)) #create figure and axis object

x = df[df['user_type']=='Subscriber']['end_station_name'].value_counts().iloc[:5]
y = df[df['user_type']=='Customer']['end_station_name'].value_counts().iloc[:5]
ax1.bar(x.index, x.values,width=.4) 
ax1.tick_params(axis='x', rotation=45)
ax2.bar(y.index, y.values,width=.4) 
ax2.tick_params(axis='x', rotation=45)
# set the title
fig.suptitle('Most popular end station, Subscribers[left] vs Customers[right]')

# show the chart
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(16,6)) #create figure and axis object

# create temp column for the viz
df['start_time'] = pd.to_datetime(df['start_time']) #convert to datetime
df['hour'] = df['start_time'].dt.hour #extract hour 

x = df[df['user_type']=='Subscriber']['hour'].value_counts().iloc[:5] 
y = df[df['user_type']=='Customer']['hour'].value_counts().iloc[:5]

ax1.bar(x.index, x.values) 
ax2.bar(y.index, y.values) 

# set the title
fig.suptitle('Rush hours, Subscribers[left] vs Customers[right]')

# show the chart
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(16,6))

x1 = df[df['user_type']=='Subscriber']['gender'].value_counts()
y1 = df[df['user_type']=='Customer']['gender'].value_counts()

ax1.bar(x1.index,x1.values,width=.4)
ax2.bar(y1.index,y1.values,width=.4)

fig.suptitle('Users Demographics(gender), Subscribers[left],Customers[right]')

plt.show()
```

```python
fig, ax1 = plt.subplots()

# create temp column for the viz
df['age'] = 2019 - df['birth_year']
x = df.groupby('user_type').mean()

ax1.bar(x.index,x['age'],width=.4)

fig.suptitle('Users Demographics(average age)')

plt.show()
```
